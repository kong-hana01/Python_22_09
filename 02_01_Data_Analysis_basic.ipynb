## Data Analysis Basic
### 데이터 분석이란?
기업의 의사결정을 위해 (빅)데이터를 정리/변환/모델링의 과정을 거쳐 데이터를 분석하는 것을 의미합니다.<br>
일반적으로 <b>문제 정의 -> 데이터 수집 -> 데이터 전처리와 EDA -> 모델링 -> 의사결정</b>의 과정을 거칩니다.<br>
아주 간단한 데이터 분석 과정을 배워보도록 하겠습니다.<br>

### 문제 정의
기업의 현재 문제를 어떤식으로 정의하느냐에 따라 앞으로 진행하게될 데이터 분석의 방향이 달라집니다.<br>
cf) 오티스사 엘레베이터 거울

- Top-down 방식: 기업 임원진들이 문제를 확인하고 분석하라고 하는 방식
- Bottom-up 방식: 실무진이 데이터를 확인하고 데이터 분석 후 임원진들에게 보고하는 방식


### 데이터 수집
앞서 정의한 문제와 관련된 질좋은 데이터 수집 과정이 필요합니다.<br>
실제로 질좋은 데이터가 없어서 데이터 분석 과정이 매우 어려울 수 있습니다.<br>
cf) Garbage in, Garbage out

- 추천 사이트: https://www.data.go.kr/   https://www.kaggle.com/


### 데이터 전처리와 EDA
데이터 전처리와 EDA(Exploratory Data Analysis) 과정을 거쳐 현재 주어진 데이터를 확인하고 정제하는 작업을 진행합니다.<br>
전처리 이후에 EDA를 진행하기에 이 둘을 구분하는 경우도 있지만<br>
EDA를 통해 데이터의 모양이 이상한 것을 확인한다면 다시 전처리를 진행해야합니다.<br>
데이터 전처리란 데이터 분석(모델링) 방향에 맞도록 데이터를 정제하는 작업을 의미합니다.<br>
실제 데이터 분석에서 70% 이상을 차지하기도 합니다.<br>
EDA란 데이터 분석에 앞서 데이터의 시각화를 통해 데이터에 대해 이해하는 과정을 의미합니다.<br>

### 모델링
실제 데이터를 머신러닝을 통해 학습시켜 모델을 만드는 과정입니다.<br>
일반적으로 훈련집합과 테스트집합(+검증집합)으로 나눠 학습과 테스트를 진행하여 모델의 정확도를 구합니다.<br>
데이터 분석은 크게 두 가지로 분류할 수 있습니다.<br>
- 예측: 여태까지의 데이터를 토대로 미래의 일을 예측하는 것 ex) 주식 가격 예측 / 농산물 가격 예측 등
- 분류: 여태까지의 데이터를 토대로 추후에 주어진 데이터를 분류하는 것 ex) 신용불량자 분류 / 암 환자 분류 등


### 의사결정
분석된 결과를 바탕으로 실제 의사결정에 반영하는 것입니다.<br>
cf) 월마트 맥주와 기저귀 사례 -> 연관 분석(장바구니 분석)


### 데이터 분석 예시
- Titanic Survival Prediction
- Iris Analysis
- 은행 계좌 정보를 토대로 신용불량자 분류
- 교통사고 다발지역 분류
- 서울시 지하철 혼잡도 분석을 통한 지하철 노선도 제안 등등
